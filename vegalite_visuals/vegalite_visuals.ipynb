```python
# Boxplot

# Box plot is used to give visuals about where 50% of the data lies, but also
# Minimum, First Quartile, Median (Second Quartile), Third Quartile and Maximum

VegaLite({
  "width": 450,
  "height": 360,
  "$schema": "https://vega.github.io/schema/vega-lite/v3.json",
  "mark": {
      "type":"boxplot",
      "extent": 1.5   # By default constant is 1.5. No need to specify unless required.
  },
  "encoding": {
      "y": {"field": "HP" , "type": "quantitative" },
#      "size": {"value": 40}
  }
}, df)
```

```python
#Pandas's melt() function. It takes 3 arguments:

#First, the DataFrame to melt.
#Second, ID variables to keep (Pandas will melt all of the other ones).
#Finally, a name for the new, melted variable.

melted_df = pd.melt(df, 
                    id_vars=["Name", "Type 1", "Type 2"], # Variables to keep
                    var_name="Stat") # Name of melted variable

# All 6 of the stat columns have been "melted" into one,
# and the new Stat column indicates the original stat 
# (HP, Attack, Defense, Sp. Attack, Sp. Defense, or Speed)

melted_df.head()
```

```python
# Boxplot from multiple columns are created by manipulating dataframe.
# all the columns are melted into two columns - Stat (contains column name)
# & value (contains column values).
VegaLite({
  "width": 450,
  "height": 360,
  "$schema": "https://vega.github.io/schema/vega-lite/v3.json",
  "mark": {
      "type":"boxplot",
      "extent": 1.5   # By default constant is 1.5. No need to specify unless required.
  },
  "encoding": {
      "x": {"field": "Stat" , "type": "nominal" },
      "y": {"field": "value" , "type": "quantitative" }
  }
}, melted_df)
```

```python
#Pandas's melt() function. It takes 3 arguments:

#First, the DataFrame to melt.
#Second, ID variables to keep (Pandas will melt all of the other ones).
#Finally, a name for the new, melted variable.

stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)

# Above in drop syntax, axis means whether to drop labels from the index (0 or index)
# or columns (1 or columns).

melted_df = pd.melt(stats_df, 
                    id_vars=["Name", "Type 1", "Type 2"], # Variables to keep
                    var_name="Stat") # Name of melted variable

# All 6 of the stat columns have been "melted" into one,
# and the new Stat column indicates the original stat 
# (HP, Attack, Defense, Sp. Attack, Sp. Defense, or Speed)

melted_df.head()
```

```python
# Boxplot from multiple columns are created by manipulating dataframe.
# all the columns are melted into two columns - Stat (contains column name)
# & value (contains column values).
VegaLite({
  "width": 450,
  "height": 300,
  "$schema": "https://vega.github.io/schema/vega-lite/v3.json",
  "mark": {
      "type":"boxplot",
      "extent": 1.5   # By default constant is 1.5. No need to specify unless required.
  },
  "encoding": {
      "x": {"field": "Stat" , "type": "nominal" },
      "y": {"field": "value" , "type": "quantitative" },
      "size": {"value": 40}
  }
}, melted_df)
```
